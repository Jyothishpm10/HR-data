```python
# import numpy
import numpy as np
import pandas as pd
#import seaborn for visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create normaldistribution of sample size 1000
sample = np.random.normal(loc = 165, scale = 10,size = 1000)
mean = sample.mean()
std = sample.std()
```

```python
plot = sns.displot(sample, kde = True)
plt.axvline(mean,ls = '-.', color = 'red')
y = plt.yticks()
plt.annotate(f'mean :{round(mean,2)}',(mean +2,110), color = 'red')
plt.axvline(mean+std,color = 'orange')
plt.annotate(f'std :{round(std,2)}',(mean+std +2,90), color = 'orange')
plt.axvline(mean-std,color = 'orange')
plt.axvline(mean+(2*std),color = 'yellow')
plt.axvline(mean-(2*std),color = 'yellow')
plt.axvline(mean+(3*std),color = 'green')
plt.axvline(mean-(2*std),color = 'green')
plt.show()
```
